## Task 3

```python
numeric_vars = ["age", "duration", "campaign", "pdays"]

for var in numeric_vars:
    plt.figure(figsize=(5, 3))
    plt.hist(df[var].dropna(), bins=30)
    plt.title(f"Histogram of {var}")
    plt.xlabel(var)
    plt.ylabel("Frequency")
    plt.show()
```
