## Task 3

```python
# define numeric variables
numeric_vars = ["age", "duration", "campaign", "pdays"]
```

### Task 3A

```python
fig, axes = plt.subplots(nrows=1,
                         ncols=len(numeric_vars),
                         figsize=(20,5))

for ax, var in zip(axes, numeric_vars):
    df[var].hist(bins=25, ax=ax)
    ax.set_title(f"Distribution of {var}", fontsize=14)
    ax.set_xlabel(var)
    ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

### Task 3B

```python
# boxplots
for var in numeric_vars:
    plt.figure(figsize=(7, 3))
    plt.boxplot(df[var].dropna(), vert=False)
    plt.title(f"Boxplot of {var}")
    plt.xlabel(var)
    plt.show()
```

### Task 3C

```python
# deciles
deciles = np.linspace(0, 1, 11)

for var in numeric_vars:
    print(f"\nDeciles for {var}:")
    print(df[var].quantile(deciles))
```

## Task 4

```python
# define categorical variables
categorical_vars = ["y", "day_of_week", "loan", "marital"]
```

### Task 4A

```python
# counts and proportions
for var in categorical_vars:
    counts = df[var].value_counts(dropna=False)
    props = df[var].value_counts(normalize=True, dropna=False)

    summary = counts.to_frame("count")
    summary["proportion"] = props

    print(f"\nSummary for {var}:")
    print(summary)
```

### Task 4B

```python
for var in categorical_vars:
    counts = df[var].value_counts(dropna=False)

    plt.figure(figsize=(8,5))
    sns.barplot(x=counts.index, y=counts.values)

    plt.title(f"Counts of {var}", fontsize=14)
    plt.xlabel(var)
    plt.ylabel("Count")
    plt.show()
```
